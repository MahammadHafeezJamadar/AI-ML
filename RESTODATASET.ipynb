```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Restaurant customer data.csv")
colors = ['Crimson' if w > 70 else 'CornflowerBlue' for w in df['weight']]

df.tail()
```

```python
about = df['marital_status'].value_counts()
plt.bar(about.index,about.values, color=["SaddleBrown","LightSalmon","blue","lightblue"])
plt.title("Marital Status Distribution")
plt.xlabel("Marital Status")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
plt.scatter(df['weight'], df['height'], c=colors)
plt.title("Weight vs Height")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.grid(True)
plt.show()
```

```python
plt.hist(df['birth_year'],bins=5,color='DarkSeaGreen',edgecolor='black')
plt.title("Birth Year Distribution")
plt.xlabel("Birth Year")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
about = df['activity'].value_counts()
plt.bar(about.index,about.values, color=["SlateBlue","red","blue","lightblue","brown"])
plt.title("Activity Distribution")
plt.xlabel("Activity")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
plt.scatter(df['birth_year'],df['weight'],c=colors)
plt.title("Birth Year vs Weight")
plt.xlabel("Birth Year")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['smoker'],bins=10,color="lightgreen",edgecolor="black")
plt.title("Distribution of Smokers")
plt.xlabel("Smokers")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
plt.hist(df['budget'],bins=10,color='MediumSeaGreen',edgecolor='black')
plt.title("Budget Distribution")
plt.xlabel("Budget")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```
